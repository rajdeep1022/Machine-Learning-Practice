```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('pokedex.csv')
dataset.head()
```

```python
dataset.isnull().sum()
```

**Finding Outliers**

```python
dataset.describe()
```

```python
sns.boxplot(x = "height_m",data=dataset)
```

```python
sns.displot(dataset['height_m'],kde=True)
```

```python
dataset.shape
```

**3-sigma rule**

```python
min_range = dataset['height_m'].mean() - (3*dataset['height_m'].std())
max_range = dataset['height_m'].mean() + (3*dataset['height_m'].std())
```

```python
min_range,max_range
```

```python
new_data = dataset[dataset["height_m"]<=max_range]
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x = "height_m",data=new_data)
plt.show()
```

**OUTLIER REMOVAL USING Z - SCORE**

```python
z_score = (dataset["height_m"] - dataset["height_m"].mean())/(dataset["height_m"].std())
```

```python
z_score
```

```python
z_score>3
```

```python
dataset["z_score"] = z_score
```

```python
dataset[dataset["z_score"]<3]
```
