```python
import pandas as pd
```

```python
df = pd.DataFrame({"Size":["s","m","l","s","s","l","xl","m"]})
df
```

Ordinal Encoding By scikit-learn

for deciding order

```python
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
df["Size_en"] = oe.fit_transform(df[["Size"]])
```

```python
df
```

Ordinal Encoding By Map function

```python
ord_data1 = {"s":0,"m":1,"l":2,"xl":3}
```

```python
df["Size_en_map"] = df["Size"].map(ord_data1)
```

```python
df
```

in map function we can give rank according to us there is no rule that it should be start from 0 . it can be customized

```python
ord_data2 = {"s":5,"m":6,"l":4,"xl":3}
```

```python
df["Size_en_map2"] = df["Size"].map(ord_data2)
```

```python
df
```
