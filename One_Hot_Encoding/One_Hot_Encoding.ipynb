```python
import pandas as pd
```

```python
dataset = pd.read_csv("College_Admission.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
en_data = dataset[["admission_status","scholarship_eligibility"]]
en_data
```

pd.get_dummies() is a pandas function used to perform One Hot Encoding.

ðŸ‘‰ It converts categorical (text) columns into binary (0/1) columns.

ðŸ‘‰ It creates a separate column for each category.

```python
pd.get_dummies(en_data)
```

```python
pd.get_dummies(en_data).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

fit() learns the unique categories in your data. and trasform() converts them into one-hot vectors.

By default, OneHotEncoder returns a sparse matrix.toarray() converts it to a normal NumPy array

```python
ohe = OneHotEncoder()
ar = ohe.fit_transform(en_data).toarray()
ar
```

```python
pd.DataFrame(ar,columns=["admission_status_admitted","admission_status_rejected","scholarship_eligibility_no","scholarship_eligibility_yes"])
```

drop="first" removes the first dummy column to avoid dummy variable trap.

```python
ohe = OneHotEncoder(drop="first")
arr = ohe.fit_transform(en_data).toarray()
arr
```

```python
pd.DataFrame(arr,columns=["admission_status_rejected","scholarship_eligibility_yes"])
```
